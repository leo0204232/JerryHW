# Q3.位移取值

```python
import requests
```

```python
response = requests.get("http://www.novelscape.net/wxxs/j/jingyong/ffwz/001.htm")
```

```python
response.encoding = "big5"
```

```python
book1 = response.text
```

```python
print('book1='+book1)
```

```python
print('[ : : ]--->'+book1[:])
```

```python
print('[10: : ]--->'+book1[10:])
```

```python
print('[ :10: ]--->'+book1[:10])
```

```python
print('[10:20: ]--->'+book1[10:20])
```

```python
print('[10:20:2]--->'+book1[10:20:2])
```

```python
print('[ :20:2]--->'+book1[:20:2])
```

```python
print('[10: :2]--->'+book1[10::2])
```

```python
print('[ : :2]--->'+book1[::2])
```
